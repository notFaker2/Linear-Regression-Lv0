```python
import pandas as pd

data=pd.read_csv('/content/drive/My Drive/ML/Linear Regression/dataset/data.csv')
display(data.head())
display(data.columns)
```

```python
columns=['X']
features=data[columns] #predictor
target=data['Y'] #to be predicted
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.5,random_state=42)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train,y_train,color='red',label='Actual')
plt.plot(X_test,y_pred,color='blue',label='Predicted')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

```python
print("Intercept(c):",model.intercept_)
print("Coefficient(m):",model.coef_)
```
